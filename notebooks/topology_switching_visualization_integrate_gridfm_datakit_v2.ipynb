# PowerGridReconfig - topology switching visualization

Этот notebook визуализирует текущий pipeline ветки `integrate-gridfm-datakit`:

1. загрузка аварийных сценариев из `gridfm-datakit`;
2. анализ перегрузок ветвей и напряжений;
3. анализ transition dataset;
4. выбор лучшего topology switching action;
5. пересчет power flow через PYPOWER backend;
6. визуальное сравнение сети **до** и **после** переключения.

Notebook рассчитан на запуск из корня репозитория или из папки `notebooks/`.

## 7. Визуализация графа сети до и после switching action

Этот блок показывает не просто граф сети, а **что именно делает выбранное действие**.

Что смотреть:

1. **Таблица Selected switching action** - какая ветвь выбрана для отключения:
   - `branch_id` - ID ветви из GridFM/MATPOWER;
   - `branch_pos` - позиция этой ветви в массиве `branch_features`;
   - `from_bus -> to_bus` - между какими узлами проходит ветвь;
   - `loading_before_percent` - загрузка этой ветви до действия.

2. **Граф BEFORE** - состояние сети до действия:
   - серая линия - нормальная ветвь, загрузка ниже 100%;
   - оранжевая линия - перегрузка 100-120%;
   - красная линия - жесткая перегрузка выше 120%;
   - черная толстая пунктирная линия - ветвь, которую мы собираемся отключить.

3. **Граф AFTER** - состояние после действия:
   - активные ветви показаны теми же цветами по загрузке;
   - красная пунктирная линия показывает, где была отключенная ветвь.

Важно: если между двумя узлами есть параллельные линии, на общей схеме они могут визуально накладываться. Поэтому точный `branch_id` всегда смотри в таблице.


````python
def get_branch_info(state, branch_id):
    """
    Return readable information about one branch in a GridFMState.

    This helps explain exactly which line/branch the topology switching action
    is going to disconnect.
    """
    if branch_id is None:
        return None

    loading_idx = BRANCH_FEATURE_COLUMNS.index("loading_percent")
    status_idx = BRANCH_FEATURE_COLUMNS.index("br_status")

    branch_positions = np.where(state.branch_ids == int(branch_id))[0]

    if len(branch_positions) == 0:
        raise ValueError(f"Branch ID {branch_id} not found in this state.")

    branch_pos = int(branch_positions[0])

    return {
        "branch_id": int(branch_id),
        "branch_pos": branch_pos,
        "from_bus": int(state.edge_index[0, branch_pos]),
        "to_bus": int(state.edge_index[1, branch_pos]),
        "status": float(state.branch_features[branch_pos, status_idx]),
        "loading_before_percent": float(state.branch_features[branch_pos, loading_idx]),
    }


selected_branch_info = get_branch_info(state_before, BRANCH_ID)

if selected_branch_info is None:
    print("Selected action is do_nothing, so no branch will be disconnected.")
else:
    print("Selected switching action:")
    display(pd.DataFrame([selected_branch_info]))


def build_visual_graph(state):
    """
    Convert GridFMState to a NetworkX graph for visualization.

    Notes:
    - only active branches are drawn as normal edges;
    - parallel branches are collapsed visually;
    - if several active branches connect the same pair of buses, the visually
      displayed edge keeps the highest loading among them.
    """
    loading_idx = BRANCH_FEATURE_COLUMNS.index("loading_percent")
    status_idx = BRANCH_FEATURE_COLUMNS.index("br_status")

    graph = nx.Graph()
    graph.add_nodes_from(range(state.bus_features.shape[0]))

    for pos_i, branch_id in enumerate(state.branch_ids):
        status = float(state.branch_features[pos_i, status_idx])
        if status <= 0:
            continue

        u = int(state.edge_index[0, pos_i])
        v = int(state.edge_index[1, pos_i])
        loading = float(state.branch_features[pos_i, loading_idx])

        edge_data = {
            "branch_id": int(branch_id),
            "loading_percent": loading,
            "branch_pos": int(pos_i),
        }

        if graph.has_edge(u, v):
            # Keep the largest loading among visually collapsed parallel edges.
            if loading > graph[u][v]["loading_percent"]:
                graph[u][v].update(edge_data)
        else:
            graph.add_edge(u, v, **edge_data)

    return graph


def edge_style(loading):
    """
    Visual style by branch loading.
    """
    if loading >= 120.0:
        return "red", 2.6
    if loading >= 100.0:
        return "orange", 2.1
    return "lightgray", 0.8


G_before = build_visual_graph(state_before)
G_after = build_visual_graph(state_after)

# Use the before-graph layout for both plots so the comparison is readable.
pos = nx.spring_layout(G_before, seed=42, iterations=80)
````

````python
from matplotlib.lines import Line2D


def draw_removed_or_highlighted_branch(ax, pos, branch_info, color, linestyle, linewidth, label):
    """
    Draw a special overlay line for the selected branch.

    We draw it manually with matplotlib so it remains visible even if the
    active graph has collapsed parallel lines.
    """
    if branch_info is None:
        return

    u = branch_info["from_bus"]
    v = branch_info["to_bus"]

    if u not in pos or v not in pos:
        return

    x1, y1 = pos[u]
    x2, y2 = pos[v]

    ax.plot(
        [x1, x2],
        [y1, y2],
        color=color,
        linestyle=linestyle,
        linewidth=linewidth,
        zorder=5,
        label=label,
    )

    # Add a small label near the middle of the selected branch.
    xm = (x1 + x2) / 2
    ym = (y1 + y2) / 2

    ax.text(
        xm,
        ym,
        f"br {branch_info['branch_id']}",
        fontsize=9,
        fontweight="bold",
        bbox={"boxstyle": "round,pad=0.2", "facecolor": "white", "edgecolor": color, "alpha": 0.9},
        zorder=6,
    )


def draw_grid_graph(
    graph,
    pos,
    title,
    selected_branch_info=None,
    mode="before",
):
    """
    Draw one network state.

    mode="before":
        black dashed overlay = branch selected for switching off.

    mode="after":
        red dashed overlay = the branch that was removed by the action.
    """
    edge_colors = []
    edge_widths = []

    for _, _, data in graph.edges(data=True):
        color, width = edge_style(data["loading_percent"])
        edge_colors.append(color)
        edge_widths.append(width)

    fig, ax = plt.subplots(figsize=(13, 9))

    nx.draw_networkx_nodes(
        graph,
        pos,
        node_size=38,
        node_color="white",
        edgecolors="black",
        linewidths=0.5,
        ax=ax,
    )

    nx.draw_networkx_edges(
        graph,
        pos,
        edge_color=edge_colors,
        width=edge_widths,
        alpha=0.88,
        ax=ax,
    )

    if mode == "before":
        draw_removed_or_highlighted_branch(
            ax=ax,
            pos=pos,
            branch_info=selected_branch_info,
            color="black",
            linestyle="--",
            linewidth=3.2,
            label="selected branch to switch off",
        )
    elif mode == "after":
        draw_removed_or_highlighted_branch(
            ax=ax,
            pos=pos,
            branch_info=selected_branch_info,
            color="red",
            linestyle="--",
            linewidth=3.2,
            label="branch switched off",
        )

    legend_elements = [
        Line2D([0], [0], color="lightgray", lw=2, label="<100% loading"),
        Line2D([0], [0], color="orange", lw=2, label="100-120% loading"),
        Line2D([0], [0], color="red", lw=2, label=">120% loading"),
    ]

    if selected_branch_info is not None:
        if mode == "before":
            legend_elements.append(
                Line2D([0], [0], color="black", lw=3, linestyle="--", label="selected to switch off")
            )
        else:
            legend_elements.append(
                Line2D([0], [0], color="red", lw=3, linestyle="--", label="switched off branch")
            )

    ax.legend(handles=legend_elements, loc="upper right")

    ax.set_title(title)
    ax.axis("off")
    plt.tight_layout()
    plt.show()


draw_grid_graph(
    G_before,
    pos,
    title=f"BEFORE action | scenario {SCENARIO_ID}",
    selected_branch_info=selected_branch_info,
    mode="before",
)

draw_grid_graph(
    G_after,
    pos,
    title=f"AFTER action | scenario {SCENARIO_ID} | switched off branch {BRANCH_ID}",
    selected_branch_info=selected_branch_info,
    mode="after",
)
````

## 8. Какие ветви были наиболее перегружены

````python
def top_loaded_branches(state, n=15):
    loading_idx = BRANCH_FEATURE_COLUMNS.index("loading_percent")
    status_idx = BRANCH_FEATURE_COLUMNS.index("br_status")

    rows = []
    for pos, branch_id in enumerate(state.branch_ids):
        status = float(state.branch_features[pos, status_idx])
        if status <= 0:
            continue

        rows.append(
            {
                "branch_id": int(branch_id),
                "branch_pos": int(pos),
                "from_bus": int(state.edge_index[0, pos]),
                "to_bus": int(state.edge_index[1, pos]),
                "loading_percent": float(state.branch_features[pos, loading_idx]),
            }
        )

    return pd.DataFrame(rows).sort_values("loading_percent", ascending=False).head(n)


print("Top loaded branches before action:")
display(top_loaded_branches(state_before, n=15))

print("Top loaded branches after action:")
display(top_loaded_branches(state_after, n=15))
````
